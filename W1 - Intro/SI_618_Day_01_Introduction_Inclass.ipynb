### <font color="magenta">Q5: (16 points)</font>
Complete the implementation of the function below.  Use the function to make plots for as many names as you like but create at least three. Then, identify at least three interesting names and explain, using complete sentences, why they're interesting.

````python
def plot_trend(name, gender):
    temp = df_names[(df_names.name == name) & (df_names.gender == gender)]
    temp.plot('year', 'birth_count', title=name)
````

````python
plot_trend('Nate', 'M')
plot_trend('Nathan', 'M')
plot_trend('Nathaniel', 'M')
````

"Nathan" and "Nathaniel" seem to have a similar lifecycle which spiked in the 80s and peaked in the 2000s. Whereas "Nate" peaked and spiked in the 2000s. The caveat is that there are very few "Nate"s overall compared to the others.
